```python
import matplotlib.pyplot as plt

plt.scatter([1, 5], [3, 4])
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0, 7)
plt.ylim(0, 5)

plt.annotate("$(x_1, y_1) = (1, 3)$", (0.5, 3.25))
plt.annotate("$(x_2, y_2) = (4, 4)$", (4.5, 4.25))
```

```python

import numpy as np
import sklearn.linear_model


X = [i for i in range(1, 6)]
Y = [2, 4, 3, 4, 5]

```

```python
plt.scatter(X, Y)
```

```python
m = sklearn.linear_model.LinearRegression()
```

```python
m
```

```python
X = np.array(X).reshape(-1, 1)
X
```

```python
m.fit(X, Y)
```

```python
plt.scatter(X, Y, label='data')
plt.plot(X, m.predict(X), label='model')
plt.legend()
```
